```python
import pandas as pd
import numpy as np
import math
import json
import copy
import statistics as sc
```

```python
def getAccuracyMean(path):
    file = open(path)
    data = json.load(file)
    count = 0
    result = []
    optimizeds = 0
    equalDefault = 0
    
    best_means = 0
    best_modes = 0
    best_good_means = 0
    best_good_modes = 0
    directlys = 0
    for i in data:
        nums = [i['best_mean'], i['best_mode'], i['best_mean_good_datasets'], i['best_mode_good_datasets'], i['best_result_using_hyperparameter_directly']]
        if(i['default_result'] < max(nums)):
            result.append(max(nums) - i['default_result'])
            optimizeds += 1
        elif(i['default_result'] == max(nums)):
            equalDefault += 1
            
        if(i['best_mean'] > i['default_result']):
            best_means += 1
            
        if(i['best_mode'] > i['default_result']):
            best_modes += 1
            
        if(i['best_mean_good_datasets'] > i['default_result']):
            best_good_means += 1
            
        if(i['best_mode_good_datasets'] > i['default_result']):
            best_good_modes += 1
        
        if(i['best_result_using_hyperparameter_directly'] > i['default_result']):
            directlys += 1

        
    print('Path: ', path)
    print('Mean: ', sc.mean(result))
    print('Median: ', sc.median(result))
    print('Max: ', max(result))
    print('Min: ', min(result))
    print('Stdev: ', sc.stdev(result))
    print('Total Optimizeds: ', optimizeds)
    print('Equal default: ', equalDefault )
    print('Improved Datasets with mean: ', best_means)
    print('Improved Datasets with mode: ', best_modes)
    print('Improved Datasets with good means: ', best_good_means)
    print('Improved Datasets with good modes: ', best_good_modes)
    print('Improved Datasets directly: ', directlys)
    print('--------------------')

    return (path, sc.mean(result))

results = []

(getAccuracyMean('final_result_general.json'))
(getAccuracyMean('final_result_statistical.json'))
(getAccuracyMean('final_result_infoTheory.json'))
(getAccuracyMean('final_result_general_infoTheory.json'))
(getAccuracyMean('final_result_statistical_infoTheory.json'))
(getAccuracyMean('final_result_general_statistical.json'))
(getAccuracyMean('final_result_general_statistical_infoTheory.json'))

# for i in results:
#     print(i[0], i[1])
```
